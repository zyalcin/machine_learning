```python
def categorise(row):  
    if row['Adjusted_Total'] >= 0 and row['Adjusted_Total'] <= 199:
        return 0
    elif row['Adjusted_Total'] > 200 and row['Adjusted_Total'] <= 600:
        return 1
```

```python
df3['discrete'] = df3.apply(lambda row: categorise(row), axis=1)
df3
```
